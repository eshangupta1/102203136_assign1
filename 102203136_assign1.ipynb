```python

def topsis(data, weights, impacts):
   
    # Normalize the decision matrix
    norm_data = data / np.sqrt((data**2).sum(axis=0))
    
    # Apply weights to the normalized decision matrix
    weighted_data = norm_data * weights
    
    # Determine ideal best and ideal worst
    ideal_best = np.where(np.array(impacts) == '+', weighted_data.max(axis=0), weighted_data.min(axis=0))
    ideal_worst = np.where(np.array(impacts) == '+', weighted_data.min(axis=0), weighted_data.max(axis=0))
    
    # Calculate distances from ideal best and worst
    distance_best = np.sqrt(((weighted_data - ideal_best)**2).sum(axis=1))
    distance_worst = np.sqrt(((weighted_data - ideal_worst)**2).sum(axis=1))
    
    # Calculate TOPSIS score
    score = distance_worst / (distance_best + distance_worst)
    
    # Rank the alternatives based on scores
    rank = score.argsort()[::-1] + 1
    
    # Return the result with scores and ranks
    result = pd.DataFrame({'TOPSIS Score': score, 'Rank': rank})
    return result

# Main program logic
#if len(sys.argv) != 5:
    print("Usage: python program.py <InputDataFile> <Weights> <Impacts> <ResultFileName>")
    print("Example: python program.py data.csv \"1,1,1,1,1\" \"+,+,+,+,-\" result.csv")
# else:
try:
    
    input_file ="C:/Users/Admin/Downloads/102203136-data.xlsx"
    weights="1,1,1,1,1"
    impacts="+,+,+,+,-" 
    result_file="C:/Users/Admin/Downloads/102203136-result.csv"
    
    # Check if file exists
    if not os.path.exists(input_file):
        print(f"Error: File '{input_file}' not found.")
        sys.exit(1)
    
    # Read the input file
    data = pd.read_csv(input_file) if input_file.endswith('.csv') else pd.read_excel(input_file)
    
    # Check if the file contains at least three columns
    if data.shape[1] < 3:
        print("Error: Input file must contain at least three columns.")
        sys.exit(1)
    
    # Parse weights and impacts
    weights = list(map(float, weights.split(',')))
    impacts = impacts.split(',')
    
    # Validate weights and impacts
    if len(weights) != len(impacts) or len(weights) != (data.shape[1] - 1):
        print("Error: Number of weights and impacts must match the number of criteria (excluding the identifier column).")
        sys.exit(1)
    if not all(i in ['+', '-'] for i in impacts):
        print("Error: Impacts must only contain '+' or '-' values.")
        sys.exit(1)
    
    # Exclude the first column (assumed to be identifiers) from the decision matrix
    decision_matrix = data.iloc[:, 1:].values
    
    # Perform TOPSIS analysis
    result = topsis(decision_matrix, weights, impacts)
    
    # Combine the results with identifiers
    final_result = pd.concat([Data.iloc[:, 0], result], axis=1)
    print("\nTOPSIS Results:")
    print(result)
    final_result = pd.concat([data.iloc[:, 0], result], axis=1)
    
    # Save the results to a CSV file
    final_result.to_csv(result_file, index=False)
    print(f"Results saved successfully to '{result_file}'")

except Exception as e:
    print(f"Error: {e}")
```
